## **Velocity Study**

```python
velocity_study = performance_metrics["velocity_study"]
speed_ranges = list(velocity_study.keys())
average_reward = [v["average_reward"] for v in velocity_study.values()]
average_steps_to_merge = [v["average_steps_to_merge"] for v in velocity_study.values()]
successful_merges = [v["successful_merges"] for v in velocity_study.values()]
collisions = [v["collisions"] for v in velocity_study.values()]
dangerous_episodes = [v["dangerous_episodes"] for v in velocity_study.values()]
average_episode_time = [v["average_episode_time"] for v in velocity_study.values()]

base_model_merging = performance_metrics["base_models"]["merging_model"]
base_model_reward = base_model_merging["average_reward"]
base_model_steps_to_merge = base_model_merging["average_steps_to_merge"]
base_model_successful_merges = base_model_merging["successful_merges"]
base_model_collisions = base_model_merging["collisions"]
base_model_dangerous_episodes = base_model_merging["dangerous_episodes"]
base_model_episode_time = base_model_merging["average_episode_time"]

plt.figure(figsize=(12, 6))
plt.plot(speed_ranges, average_reward, marker='o', label='average_reward')
plt.plot(speed_ranges, average_steps_to_merge, marker='o', label='average_steps_to_merge')
plt.plot(speed_ranges, successful_merges, marker='o', label='successful_merges')
plt.plot(speed_ranges, collisions, marker='o', label='Collisions')
plt.plot(speed_ranges, dangerous_episodes, marker='o', label='Dangerous Episodes')
plt.plot(speed_ranges, average_episode_time, marker='o', label='average_episode_time')
plt.axhline(base_model_reward, color='r', linestyle='--', label='Base Model Reward')
plt.axhline(base_model_steps_to_merge, color='g', linestyle='--', label='Base Model Steps to Merge')
plt.axhline(base_model_successful_merges, color='b', linestyle='--', label='Base Model Successful Merges')
plt.axhline(base_model_collisions, color='purple', linestyle='--', label='Base Model Collisions')
plt.axhline(base_model_dangerous_episodes, color='orange', linestyle='--', label='Base Model Dangerous Episodes')
plt.axhline(base_model_episode_time, color='brown', linestyle='--', label='Base Model Episode Time')

plt.title('Velocity Study Metrics Over Speed Ranges')
plt.xlabel('Speed Ranges')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

In the **(10, 20)** range, the agent demonstrates a **significant improvement.** It achieves **200 successful merges and maintains zero collisions,** indicating **effective merging behavior.** The **average reward drops slightly, suggesting a trade-off between speed and reward,** but the agent remains **stable with zero dangerous driving episodes.** Compared to the **base model,** the agent likely performs **similarly or better** in terms of **successful merges and collisions,** while maintaining a **comparable average reward.**

Moving to the **(20, 30)** range, the agent's **average reward continues to drop,** but it maintains a **high success rate with 200 successful merges and zero collisions.** The **average episode time decreases, suggesting faster decision-making at these speeds.** Compared to the **base model,** the agent likely maintains a **competitive performance in terms of successful merges and collisions,** while potentially achieving a **slightly lower average reward.**

In the **(30, 40)** range, the agent's performance begins to **degrade with a rise in collisions and dangerous driving episodes,** indicating **increasing difficulty in handling higher speeds.** Compared to the **base model,** the agent likely experiences a **decline in performance,** with a **higher number of collisions and potentially fewer successful merges.**

In the **(40, 50)** range, the agent faces **significant challenges, with a sharp increase in collisions and a decrease in successful merges.** The **average reward also drops,** suggesting a **decline in overall performance.** Compared to the **base model,** the agent's performance **significantly deteriorates,** with a **much higher number of collisions and likely fewer successful merges.**

## **Ego Merging**

```python
brake_data = performance_metrics["ego_merging_brake"]
accel_data = performance_metrics["ego_merging_accel"]
base_model_data = performance_metrics["base_models"]["merging_model"]

environments = list(brake_data.keys())

metrics = [
    "average_reward",
    "average_steps_to_merge",
    "average_episode_time",
    "collisions",
    "successful_merges",
    "dangerous_episodes"
]

brake_metrics = {metric: [brake_data[env][metric] for env in environments] for metric in metrics}
accel_metrics = {metric: [accel_data[env][metric] for env in environments] for metric in metrics}
base_model_metrics = {metric: [base_model_data[metric]] * len(environments) for metric in metrics}

fig, axs = plt.subplots(len(metrics), 1, figsize=(12, 18))

for i, metric in enumerate(metrics):
    axs[i].plot(environments, brake_metrics[metric], marker='o', label='Brake')
    axs[i].plot(environments, accel_metrics[metric], marker='o', label='Accel')
    axs[i].plot(environments, base_model_metrics[metric], linestyle='--', label='Base Model', color='red')
    axs[i].set_title(f'Comparison of {metric.replace("_", " ").title()}')
    axs[i].set_ylabel(metric.replace("_", " ").title())
    axs[i].legend()
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```

These charts focus on scenarios where the ego vehicle is entering the highway and has two possible actions: braking to allow another vehicle to merge ahead, or accelerating to merge in front of the existing vehicle. Both actions are evaluated against a base model.

The first chart, showing the **average reward**, indicates that all approachesbraking, accelerating, and the base modelyield very similar results. The **base model achieves a slightly higher reward**, but the difference is negligible. This suggests that **prioritizing other vehicles' influence results in comparable outcomes** for both braking and accelerating actions, with the base model providing only marginally better results.

In the second chart, analyzing **average steps to merge**, braking and accelerating perform identically, both requiring the same number of steps. The base model is similarly aligned. This reflects that **prioritizing others' needs does not differentiate between the two actions regarding the time needed to facilitate merging.**

The third chart, examining **average episode time**, mirrors the previous findings: both braking and accelerating have nearly identical episode times, slightly longer than the base model. This indicates that **both actions are equally efficient when prioritizing other vehicles, with only minor differences from the base model.**

For **collisions**, the results are identical across all approaches, with **zero collisions recorded.** This emphasizes the **safety of prioritizing others, regardless of whether braking or accelerating is employed.**

Regarding **successful merges**, both braking and accelerating achieve the same maximum number of successful merges, slightly outperforming the base model. This highlights the **effectiveness of prioritizing others in ensuring smooth merging, with either action proving equally capable.**

Finally, the chart for **dangerous episodes** shows no dangerous episodes for either action or the base model. This reaffirms the **inherent safety of strategies focused on others' influence, regardless of the chosen action.**

In conclusion, when prioritizing other vehicles, **both braking and accelerating actions perform similarly across all metrics, achieving high safety, efficiency, and success rates.** This analysis underscores that **either action can be effectively employed to prioritize others in merging scenarios, with minimal impact on performance relative to the base model.**
